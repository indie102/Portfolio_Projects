```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_companies_of_New_Zealand'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
print(soup)
```

```python
table = soup.find_all('table')[1]
```

```python
soup.find_all('table', class_ = "wikitable sortable")[1]
```

```python
big_titles = table.find_all('th')
print(big_titles)
```

```python
big_table = [title.text.strip() for title in big_titles][:-1]
print(big_table)
```

```python
df = pd.DataFrame(columns = big_table)
df
```
